```python
def get_color(name,number):
    pal=list(sns.color_palette(palette=name,n_colors=number).as_hex())
    return pal
```

## What is Toss Decision Ratio in IPL?

```python
total =  df['TossDecision'].value_counts().sum()
```

```python
fig = px.pie(  d1, values=d1['Ratio'], names=d1['TossDecision'])
fig.update_traces(textposition='inside',
                  textinfo='percent+label',
                  sort=False)
fig.update_layout(width=500, height=550)
fig.show()
```

```python
total = (df.TossWinner==df.WinningTeam).value_counts().sum()
```

```python
toss_vs_matchwin = (df.TossWinner==df.WinningTeam).value_counts()
toss_vs_matchwin
```

In IPL we can see that most of the teams chose to field first after winning toss.

```python
tm = toss_vs_matchwin.reset_index().rename(columns={'index':'isTossWinner',0:'count'})
tm['Percentage'] = round((tm['count']/total)*100,2)
tm
```

```python
fig = px.pie(tm, values=tm['Percentage'], names=['Toss Lost','Toss Won'])
fig.update_traces(textposition='inside',
                  textinfo='percent+label',
                  sort=False)
fig.update_layout(width=500, height=550)
fig.show()
```

> overall toss be a factor to win the match but remember around 52% times teams which lost the toss have won the match due to their better performance
